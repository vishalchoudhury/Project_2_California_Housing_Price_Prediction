#  Background of Problem Statement :

The US Census Bureau has published California Census Data which has 10 types of metrics such as the population, median income, median housing price, and so on for each block group in California. The dataset also serves as an input for project scoping and tries to specify the functional and nonfunctional requirements for it.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#  1. Load the data :

```python
housingData = pd.read_excel("1553768847_housing.xlsx")
```

```python
housingData.info()
```

```python
housingData.head()
```

```python
housingData.corr()
```

```python
#From Above correelation Age, totalrooms,total_bedrooms,population household, median income and ocean_proximity will be features.
features = housingData.iloc[:,[2,3,4,5,6,7,8]]
```

```python
features.head()
```

```python
features = features.values
label = housingData.iloc[:,-1].values
```

#  2. Handle missing values :

```python
pd.DataFrame(features).info()
```

```python
from sklearn.preprocessing import Imputer
bedRoomImputer = Imputer(missing_values = np.nan, strategy = 'mean',axis=0)
features[:,[2]] = bedRoomImputer.fit_transform(features[:,[2]])
```

```python
pd.DataFrame(features).info()
```

#  3. Encode categorical data :

```python
pd.DataFrame(features[:,-1])[0].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
oceanEnoder = LabelEncoder()
features[:,-1] = oceanEnoder.fit_transform(features[:,-1])
```

```python
from sklearn.preprocessing import OneHotEncoder
oceanOHE = OneHotEncoder(categorical_features=[-1])
features = oceanOHE.fit_transform(features).toarray()
```

```python
pd.DataFrame(features).head()
```

```python
#Applying OLS for feature scaling
finalFeatures = np.append(np.ones((20640,1)), features, axis=1)
```

```python
pd.DataFrame(finalFeatures).head()
```

```python
print(finalFeatures.shape)
print(label.shape)
```

```python
import statsmodels.api as sm
model1 = sm.OLS(endog=label, exog=finalFeatures).fit()
model1.summary()
```

```python
#As the P-Values of X5 is high we will be eleminating X5 --> housing_median_age
finalFeatures = finalFeatures[:,[0,1,2,3,4,6,7,8,9,10,11]]
model1 = sm.OLS(endog=label, exog=finalFeatures).fit()
model1.summary()
```

```python
#No more feature scaling is required as it is all the K-Value is 0.
features = finalFeatures[:,1:]
```

#  5. Standardize data: 

```python
from sklearn.preprocessing import MinMaxScaler
scaleData = MinMaxScaler(feature_range=(0,1))
features = scaleData.fit_transform(features)
```

```python
pd.DataFrame(features).head()
```

#  6. Perform Linear Regression  and 4. Split the dataset : : 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lineRegScore = dict()
for i in range(1,20641):
    X_train,X_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state=i)
    model =LinearRegression()
    model.fit(X_train,y_train)
    testing_score = model.score(X_test,y_test)
    training_score = model.score(X_train,y_train)
    if testing_score > training_score:
        lineRegScore[i] = testing_score
        print("Testing score: {} Training score: {} Random state: {}".format(testing_score,training_score,i))
```

```python
max(lineRegScore.values())
```

```python
[k for k,v in lineRegScore.items() if v == 0.67397057285073]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state=928)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print("Testing Score: {} \nTraining Score: {}".format(model.score(X_test,y_test), model.score(X_train,y_train)))
```

#  Predict output for test dataset using the fitted model.

```python
y_pred = model.predict(X_test)
```

#  7. Perform Decision Tree Regression :

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
decisionTreeScore = dict()
for i in range(1,20641):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                    label,
                                                    test_size=0.2,
                                                    random_state=i)
    modelDT = DecisionTreeRegressor(max_depth=5)
    modelDT.fit(X_train,y_train)
    testing_score = modelDT.score(X_test,y_test)
    training_score = modelDT.score(X_train,y_train)
    if testing_score > training_score:
        decisionTreeScore[i] = testing_score
        print("Testing score: {} Training score: {} Random state: {}".format(testing_score,training_score,i))
```
