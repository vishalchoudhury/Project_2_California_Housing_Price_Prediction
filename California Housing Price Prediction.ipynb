# 1. Load the data 

```python
housingData = pd.read_excel("1553768847_housing.xlsx")
```

```python
features = housingData.iloc[:,0:9].values
label = housingData.iloc[:,9].values
```

# 2. Handle missing values :

```python
#Missing data in total_bedrooms
from sklearn.preprocessing import Imputer
bedroomsImpute = Imputer(missing_values=np.nan, strategy='mean', axis=0)
features[:,[4]] = bedroomsImpute.fit_transform(features[:,[4]])
```

```python
pd.DataFrame(features).info()
```

# 3. Encode categorical data :

```python
#Handling Ocean_Proximity(Non-Numeric) column
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
oceanEncoder = LabelEncoder()
features[:,-1] = oceanEncoder.fit_transform(features[:,-1])
```

```python
oceanOHE = OneHotEncoder(categorical_features=[-1])
features = oceanOHE.fit_transform(features).toarray()
```

```python
pd.DataFrame(features).head()
```

```python
#Feature Elemination (Using the Univariate Technique Annova)
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_regressiona
```

```python
selectPercentile = SelectPercentile(percentile=50, score_func=f_regression)
```

```python
selectPercentile.fit(features,label)
print(selectPercentile.get_support())
```

```python
features = features[:,[2,5,7,8,9,10,11]]
```

```python
pd.DataFrame(features).head()
```

```python
#Standardize the data set
from sklearn.preprocessing import MinMaxScaler
scaleData = MinMaxScaler(feature_range=(0,1))
```

```python
features = scaleData.fit_transform(features)
```

```python
pd.DataFrame(features).head()
```

```python
#TrainTestSplit
scoreData = dict()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
for i in range(1,20641):
    X_train,X_test,y_train,y_test = train_test_split(features,label,test_size=0.2, random_state=i)
    model = LinearRegression()
    model.fit(X_train,y_train)
    training_score = model.score(X_train,y_train)
    testing_score = model.score(X_test,y_test)
    if testing_score > training_score:
        scoreData[i] = testing_score
        if testing_score == 0.6872705515887392:
            print("Best Random state: {}".format(i))
        print("Testing {}  Training {}  Random State {}".format(testing_score,training_score,i))
```

```python
print(max(scoreData.values()))
```

```python
for k,v in scoreData.items():
    if v == 0.6872705515887392:
        print(k)
```
